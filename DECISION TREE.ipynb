```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("salaries.csv")
```

```python
data
```

```python
data.info()
```

```python
x = data.drop("salary_more_then_100k", axis = 1)
```

```python
x
```

```python
y = data["salary_more_then_100k"]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
x["company_new"] = le_company.fit_transform(x["company"])
x["job_new"] = le_job.fit_transform(x["job"])
x["degree_new"] = le_degree.fit_transform(x["degree"])
```

```python
x
```

```python
x = x.drop(["company", "job", "degree"], axis = 1)
```

```python
x
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[2,2,1]])
```

```python
model.predict([[1,1,1]])
```

##### In this file using following columns build a model to predict if person would survive or not,

1. Pclass
1. Sex
1. Age
1. Fare

```python
data_2 = data_2.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis = 1)
```

```python
data_2
```

```python
y =  data_2["Survived"]
```

```python
y
```

```python
x = data_2.drop(["Survived"], axis = 1)
```

```python
x
```

```python
x["Age"] = x["Age"].fillna(x["Age"].mean())
```

```python
x.info()
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
x["sex_new"] = le_sex.fit_transform(x["Sex"])
```

```python
x
```

```python
x = x.drop(["Sex"], axis = 1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
len(x_train)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
